### Função filtrar nota

```python
def grade_filter(subject, grade):
    print("A disciplina escolhida, foi:", subject)
    print("A nota escolhida", grade)
    print("Número de alunos: ", len(dataMerge[dataMerge[subject]>=grade]))
    return dataMerge[dataMerge[subject]>=grade]
```

```python
grade_filter("Nota_Redacao", 77)
```

```python
def grade_filter_percent(subject: str, grade: int):
    print("A disciplina escolhida, foi:", subject)
    print("A nota escolhida", grade)
    print("Número de alunos: ", len(dataMerge[dataMerge[subject]>=grade]))
    return len(dataMerge[dataMerge[subject]>grade]/len(alunos)*100)
```

```python
grade_filter_percent("Nota_Redacao", 95)
```
