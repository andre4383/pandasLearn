```python
import pandas as pd

url = "https://drive.google.com/uc?id=15aOJIGAyLMSY1gecjiCgu2ko_riIcKQy"
estudantes = pd.read_csv(url)

url2 = "https://drive.google.com/uc?id=1Jgto7psHaMRTAVzcFt7D6SgJiHMB7uGT"
escolas = pd.read_csv(url2)
```

```python
escolas
```

```python
estudantes
```

```python
data = pd.merge(estudantes, escolas, on='Nome_Escola')
```

## Roi Educacional

Fórmula = Ganhos - Custos / Custos

```python
data['Custo_Aluno'] = data['Orcamento_Anual'] / data['Numero_Alunos']
```

```python
data['Nota_Media'] = (data['Nota_Redacao'] + data['Nota_Matematica']) / 2
```

```python
data
```

```python
data['Eficiencia'] = data['Nota_Media'] / data['Custo_Aluno']
```

Eu quero mostrar quais escolas são as mais eficientes

```python
ranking_escolas = data.groupby('Nome_Escola')[['Nota_Media', 'Custo_Aluno', 'Eficiencia']].mean().reset_index()
```

```python
ranking_eficiencia = ranking_escolas.sort_values(by='Eficiencia', ascending=False)
```
